```python
gw2tp_items= {
    77750: "Lucky Draketail",
    19976: "Mystic Coin",
    82678: "Pulsing Brandspark",
    19700: "Mithril Ore",
    36041: "Piece of Candy Corn",
    19703: "Silver Ore",
    82991: "Bag of Coffee Beans",
    49424: "+1 Agony Infusion",
    19698: "Gold Ore",
    19729: "Thick Leather Section",
}
```

```python
import os
os.getcwd()
os.getcwdb()
```

```python
from datetime import datetime
from ml.common.dataRetriever.data_retriever import DataRetriever

#directory in project root.
datapath = os.path.join(os.getenv('DATA_PATH'), 'raw')
os.makedirs(datapath, exist_ok=True)
# Create retriever
retriever = DataRetriever()
for key,value in gw2tp_items.items():
    df = retriever.get_data(
        symbol=str(key),
        source="DAMU",
        start_time=datetime(2025, 5, 1),
        end_time=datetime(2025, 12, 23),
        interval="15m", # does not matter for damu
    )
    df.to_csv(f"{datapath}/{value}.csv", index=False, sep=";")
    print(f"Saved {df.count()} for {value}")
```

```python
from ml.common.folder_s3_sync import upload_folder_to_s3
import time

# Set up parameters
local_folder = datapath
unix_timestamp = int(time.time())
s3_folder_prefix = f'datasources/gw2/raw/{unix_timestamp}/' # folder on s3

# Call the function
upload_folder_to_s3(
    local_folder=local_folder,
    s3_folder_prefix=s3_folder_prefix,
    unix_timestamp=unix_timestamp,
    bucket_name="ost-s3"  # optional, defaults to "ost-s3"
)
```
