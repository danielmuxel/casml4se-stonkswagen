```python

# ... existing code ...

from datetime import datetime
from gw2ml.data.retriever import DataRetriever
from gw2ml.data.s3_sync import upload_folder_to_s3
import os
import time
import shutil

datapath = os.path.join(os.getenv('DATA_PATH'), 'gw2raw')
os.makedirs(datapath, exist_ok=True)
print("Creating data at {}".format(datapath))

# Create retriever
retriever = DataRetriever()

# Set timestamp once for all chunks
unix_timestamp = int(time.time())
s3_folder_prefix = f'datasources/gw2/raw/{unix_timestamp}/'

# Convert items dict to list of tuples for chunking
items_list = list(gw2tp_items.items())
chunk_size = 10

# Process in chunks
for chunk_idx in range(0, len(items_list), chunk_size):
    chunk = items_list[chunk_idx:chunk_idx + chunk_size]

    print(
        f"\nProcessing chunk {chunk_idx // chunk_size + 1} (items {chunk_idx + 1} to {min(chunk_idx + chunk_size, len(items_list))})")

    # Process each item in the chunk
    for key, value in chunk:
        df = retriever.get_data(
            symbol=str(key),
            source="DAMU",
            start_time=datetime(2025, 3, 1),
            end_time=datetime(2025, 11, 30),
            interval="15m",
        )
        df.to_csv(f"{datapath}/{value}.csv", index=False, sep=";")
        print(f"Saved {df.count()} for {value}")

    # Upload chunk to S3
    print(f"Uploading chunk to S3...")
    upload_folder_to_s3(
        local_folder=datapath,
        s3_folder_prefix=s3_folder_prefix,
        unix_timestamp=unix_timestamp,
        bucket_name="ost-s3"
    )

    print(f"Chunk {chunk_idx // chunk_size + 1} uploaded and local files cleared")

print(f"\nAll chunks uploaded to S3 with timestamp: {unix_timestamp}")
```
